### Exercise 2

#### Exercise 2.a

Using the OLS loss, try to learn a classifier for the dataset given below. 

```python

import scipy.io
import matplotlib.pyplot as plt

data_class1 = scipy.io.loadmat('points_class1_Lab2_Ex1.mat')['points_class1_Lab2_Ex1']
data_class2 = scipy.io.loadmat('points_class2_Lab2_Ex1.mat')['points_class2_Lab2_Ex1']

plt.scatter(data_class1[:,0], data_class1[:,1])
plt.scatter(data_class2[:,0], data_class2[:,1])
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
data = np.vstack((data_class1,data_class2)) #class1 = blue points, class2 = orange points

t_class1 = np.ones((np.shape(data_class1)[0],1))
t_class2 = 0*np.ones((np.shape(data_class2)[0],1))

t_class = np.vstack((t_class1,t_class2))

myreg = LinearRegression()
myreg.fit(data,t_class)

#limits of space
x1min = np.min(data[:,0])
x1max = np.max(data[:,0])
x2min = np.min(data[:,1])
x2max = np.max(data[:,1])

x1 = np.linspace(x1min,x1max,200)
x2 = np.linspace(x2min,x2max,200)

#generates all possible points x1_i x2_j
xx1,xx2 = np.meshgrid(x1,x2)

#feature matrix for grid
#flatten returns horizontal vectors
Xgrid = np.vstack((xx1.flatten(),xx2.flatten())).T

#on grid we can compute classification
#get weights from trained model
beta = np.zeros((3,))
beta[1:] = myreg.coef_
beta[0] = myreg.intercept_

Xgrid_tilde = np.hstack((np.ones((np.shape(Xgrid)[0],1)),Xgrid))

pred_grid = np.matmul(Xgrid_tilde,beta)

grid_class = pred_grid>1/2

from matplotlib.colors import ListedColormap
cm_bright = ListedColormap(['#FF0000','#0000FF'])

plt.contourf(xx1,xx2,grid_class.reshape(np.shape(xx1)),alpha=0.2,cmap=cm_bright)
plt.scatter(data[:,0], data[:,1], c=t_class)
plt.show()
```
